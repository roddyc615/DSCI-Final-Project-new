```r
library(tidyverse)
```

```r
players <- read_csv("data/players.csv")
players
```

```r
players_data <- players |>
    select(subscribe, played_hours, Age)|>
    filter(Age <= 40, played_hours <= 30, played_hours >0)|>
    mutate(subscribe = factor(subscribe, levels = c(TRUE, FALSE), labels = c("true", "false")))
players_data
```

```r
mean_table <- players_data |>
    summarize(mean_played_hours = mean(played_hours, na.rm = TRUE),
              mean_age = mean(Age, na.rm = TRUE))

mean_table
```

```r
options(repr.plot.height = 7, repr.plot.width = 10)
players_plot <- players_data|>
    ggplot(aes(x = Age, y = played_hours, color = subscribe))+
    geom_point()+
    labs(x="Age of Player (years)", y="Total Played Time (hours)", color="Subscribiton Status (True or False)")+
    ggtitle("Played Hours vs. Age")+
    theme(text = element_text(size = 15))
players_plot
```

```r
library(rsample)
library(recipes)
library(tidymodels)
players_split <- initial_split(players_data, prop = 0.75, strata = subscribe)
players_train <- training(players_split)
players_test <- testing(players_split)
```

```r
players_recipe <- recipe(subscribe ~ Age + played_hours, data = players_data) |>
    step_scale(all_predictors()) |>
    step_center(all_predictors())

players_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 3) |>
    set_engine("kknn") |>
    set_mode("classification")

knn_fit <- workflow() |>
    add_recipe(players_recipe) |>
    add_model(players_spec)|>
    fit(data = players_train)

knn_fit
```

```r
players_test_predictions <- predict(knn_fit, players_test) |>
    bind_cols(players_test)

players_test_predictions
```

```r
players_test_predictions |>
    metrics(truth = subscribe, estimate = .pred_class) |>
    filter(.metric == "accuracy")
```

```r
players_test_predictions |> pull(subscribe) |> levels()
```

```r
players_test_predictions |> 
    precision(truth = subscribe, estimate = .pred_class, event_level = "first")
```

```r
players_test_predictions |> 
    recall(truth = subscribe, estimate = .pred_class, event_level = "first")
```

```r
confusion <- players_test_predictions |>
    conf_mat(truth = subscribe, estimate = .pred_class)
confusion
```

```r
set.seed(123)

#CV folds on the training
player_folds <- vfold_cv(players_train, v=5, strata = subscribe)

#Make k and and distance tunable
players_tune_spec <- nearest_neighbor(
    weight_func = "optimal",
    neighbors   = tune(),
    dist_power = tune()
) %>% 
    set_engine("kknn") %>%
    set_mode ("classification")

players_wf <- workflow() %>%
  add_recipe(players_recipe) %>%
  add_model(players_tune_spec)

#Grid search
tune_res <- tune_grid(
  players_wf,
  resamples = player_folds,
  grid = grid_regular(
    neighbors(range = c(1, 25)),
    dist_power(range = c(1, 3)),
    levels = c(10, 5)
  ),
  metrics = metric_set(accuracy, roc_auc, sens, spec)
)

#Pick best
best_params <- select_best(tune_res, "accuracy")
final_wf    <- finalize_workflow(players_wf, best_params)

final_fit <- last_fit(final_wf, players_split)
collect_metrics(final_fit)
collect_predictions(final_fit) %>% conf_mat(truth = subscribe, estimate = .pred_class)
```

```r

library(ggplot2)

tune_plot <- tune_res %>%
  collect_metrics() %>%
  filter(.metric == "accuracy") %>%
  mutate(dist_power = factor(dist_power))

ggplot(tune_plot,
       aes(x = neighbors, y = mean, color = dist_power)) +
  geom_line() +
  geom_point() +
  labs(
    x = "Number of neighbors (k)",
    y = "Mean CV accuracy",
    color = "dist_power"
  ) +
  theme_minimal()
```
